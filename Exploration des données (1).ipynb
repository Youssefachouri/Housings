En utilisant des graphiques à barres, j'ai analysé la répartition des valeurs pour différentes catégories, ainsi que les décomptes des valeurs uniques pour certaines variables. Cette approche m'a permis d'identifier quelles colonnes conserver et lesquelles supprimer dans notre dataset.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('Ads.csv')
data
```

```python
value_counts = data['AGENCIES_UNWANTED'].value_counts()

sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))

sns.barplot(x=value_counts.index, y=value_counts.values)

plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Value Counts of Categories')
plt.show()
```

```python
value_counts = data['PARKING_COUNT'].value_counts()

sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))

sns.barplot(x=value_counts.index, y=value_counts.values)

plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Value Counts of Categories')
plt.show()
```
