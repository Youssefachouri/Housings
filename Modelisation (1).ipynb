```python
# Data manipulation
import pandas as pd
import numpy as np
# To import BERT-based model and measure sequence similarities
from sentence_transformers import SentenceTransformer, util
# Feature scaling
from sklearn.preprocessing import StandardScaler
# Clustering
from sklearn.cluster import DBSCAN
```

# Import data

```python
# Read and prepare data
data = pd.read_csv('Ads.csv')
data
```

```python
nan_descriptions = pd.DataFrame(columns=data.columns)

condition = data['DESCRIPTION'].apply(lambda x: not isinstance(x, str))
rows_to_append = data[condition]

nan_descriptions = pd.concat([nan_descriptions, rows_to_append])
data.drop(data[data['DESCRIPTION'].apply(lambda x: not isinstance(x, str))].index, inplace=True)
data.reset_index(inplace=True)
```

# Methodologie

La première intuition pour regrouper les observations est d'appliquer des techniques de clustering. Les techniques de base exigent que l'on indique le nombre de clusters que l'on souhaite obtenir.

Dans ce contexte, nous ne connaissons pas le nombre réel de maisons dans l'ensemble de données puisque nous avons un problème de redondance. C'est pourquoi j'ai choisi d'appliquer la technique de Clustering DBSCAN (Density-Based Spatial Clustering of Applications with Noise) basée sur la variable "DESCRIPTION", qui est la plus informative (elle résume toutes les informations relatives à la liste des maisons).

Pour obtenir nos text embeddings (la tâche consistant à transformer des données textuelles en vecteurs), j'ai utilisé le modèle [LaBSE](https://huggingface.co/sentence-transformers/LaBSE)  de [HuggingFace](https://huggingface.co/) hub, qui est basé sur le fameux modèle [BERT: Pre-training of Deep Bidirectional Transformers for Language Understanding](https://arxiv.org/abs/1810.04805) de Google Research.  Ce modèle a la même architecture que BERT, mais il a été entraîné sur <b> un corpus français</b> , ce qui est le cas de notre dataset.

Une fois que j'ai obtenu les embeddings de texte, j'ai effectué la tâche de clustering. J'ai fixé les paramètres <b>"eps"</b> (distance maximale entre deux échantillons pour que l'un soit considéré comme étant dans le voisinage de l'autre) et<b> "min_samples"</b> (nombre d'échantillons dans un voisinage pour qu'un point soit considéré comme un noyau) de DBSCAN. J'ai effectué quelques expériences avec ces valeurs afin de trouver les paramètres optimaux pour notre dataset.

```python
# Load the LaBSE SentenceTransformer
model_name = "sentence-transformers/LaBSE"
sentence_encoder = SentenceTransformer(model_name)
```

```python
# Encode all sentences
embeddings = sentence_encoder.encode(data['DESCRIPTION'].tolist())
```

```python
# Feature scaling
scaler = StandardScaler()
embeddings = scaler.fit_transform(embeddings)
data['embeddings'] = embeddings.tolist()
```

```python
# Perform clustering technique using DBSCAN
dbscan = DBSCAN(eps=0.25, min_samples=1)
dbscan.fit(embeddings)
```

```python
cluster_labels = dbscan.labels_
list(cluster_labels)
```

```python
# a function that can be used to search the indices of the observations in the cluster number n
def search(n):
  #print('These are the indices of the listings which are in the cluster n°', n,':', np.where(cluster_labels == n)[0])
  cluster_n_indices = np.where(cluster_labels == n)[0]
  most_similar_examples_in_cluster_n = data.iloc[cluster_n_indices]
  return most_similar_examples_in_cluster_n
```

```python
for i in range(2160):
  if len(search(i)) > 2:
    print(i)
```

```python
most_similar_listings = search(221)
most_similar_listings
```

```python
for i in range(len(most_similar_listings)):
  print(most_similar_listings['DESCRIPTION'].iloc[i])
```
